```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_excel('misurepython.xlsx')
df.dtypes
df
#plt.subplot(1,1,1)
plt.hist(df)
df.plot(kind='kde')
```

```python
df
```

```python
import seaborn as sns
sns.set(rc={'figure.figsize':(20,10), 'figure.facecolor':'grey'})
sns.histplot(df)
plt.xlabel('Nanometers', fontsize=30);
plt.ylabel('Count', fontsize=30);
plt.title('Distribution of target DNA', fontsize=40)
plt.tick_params(axis='both', which='major', labelsize=20)

df.plot(kind='kde',xlabel='length')
plt.xlabel('Nanometers', fontsize=30);
plt.ylabel('Density', fontsize=30);
plt.title('Distribution of target DNA', fontsize=40)
plt.tick_params(axis='both', which='major', labelsize=20)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(20,30), 'figure.facecolor':'grey'})
sns.displot(data=df, x='Estimated Length (nm)', kde=True, color='green')
plt.legend(labels=['DNA length','DNA length'], title = "Legend", fontsize = 'large', title_fontsize = "14", loc = 1, bbox_to_anchor = (1,1))
plt.xlabel('Nanometers', fontsize=16);
plt.ylabel('Count', fontsize=16);
plt.title('Distribution of target DNA', fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=15)
plt.savefig('distribuzione lunghezze DNA target.tiff', dpi=300)
```

```python
sns.set(rc={'figure.figsize':(20,10), 'figure.facecolor':'grey'})
sns.boxplot(data=df, color='green')
plt.legend(labels=['DNA length'], title = "Legend", fontsize = 'large', title_fontsize = "14", loc = 1, bbox_to_anchor = (1,1))
#plt.xlabel('Nanometers', fontsize=30);
plt.ylabel('Nanometers', fontsize=30);
plt.title('Distribution of target DNA', fontsize=40)
plt.tick_params(axis='both', which='major', labelsize=20)
plt.savefig('distribuzione lunghezze DNA target boxplot.tiff', dpi=300)
```

```python
#sns.set(rc={'figure.figsize':(10,15), 'figure.facecolor':'grey'})
sns.displot(data=df, kde=True)
plt.xlabel('Nanometers', fontsize=20);
plt.ylabel('Count', fontsize=20);
plt.title('Distribution of target DNA', fontsize=30)
plt.tick_params(axis='both', which='major', labelsize=16)
#plt.legend(labels=['DNA length','DNA length'], title = "Legend", fontsize = 'large', title_fontsize = "14", loc = 1, bbox_to_anchor = (1,1))
```

```python
sns.set(rc={'figure.figsize':(20,10), 'figure.facecolor':'grey'})
sns.swarmplot(data=df)
```

```python
sns.set(rc={'figure.figsize':(20,10), 'figure.facecolor':'grey'})
sns.stripplot(data=df, color='blue')
#sns.histplot(data=df)
sns.boxplot(data=df, color='green')
plt.legend(labels=['DNA length'], title = "Legend", fontsize = 'large', title_fontsize = "14", loc = 1, bbox_to_anchor = (1,1))
#plt.xlabel('Nanometers', fontsize=30);
plt.ylabel('Nanometers', fontsize=30);
plt.title('Distribution of target DNA', fontsize=40)
plt.tick_params(axis='both', which='major', labelsize=20)
plt.savefig('distribuzione lunghezze DNA target stripplot.tiff', dpi=300)
```

```python
sns.set_theme()

sns.set(rc={'figure.figsize':(20,10), 'figure.facecolor':'grey'})
sns.kdeplot(data=df, palette='Greens')
sns.rugplot(data=df, palette='Greens')
plt.legend(labels=['DNA length'], title = "Legend", fontsize = 'large', title_fontsize = "14", loc = 1, bbox_to_anchor = (1,1))
plt.xlabel('Nanometers', fontsize=30);
plt.ylabel('Density', fontsize=30);
plt.title('Distribution of target DNA', fontsize=40)
plt.tick_params(axis='both', which='major', labelsize=20)
plt.savefig('distribuzione lunghezze DNA target rugplot.tiff', dpi=300)
```
